```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B")
```

```python
# Add a new pad_token if it's missing
tokenizer.add_special_tokens({'pad_token': '[PAD]'})
model.resize_token_embeddings(len(tokenizer))  # Resize model's token embeddings after adding new tokens
```

```python
def generate_response(prompt):
    # Tokenize the input and apply padding/truncation
    inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True)

    # Generate the response with the attention mask
    outputs = model.generate(
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_length=100,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id  # Ensure pad_token_id is set
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

```

```python
query = "Explain the difference between finetuning LLM and RAG"
response = generate_response(query)
print(response)
```

```python
# Step 2: Extract Text from PDF
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text() + "\n"  # Add a newline for better formatting
    return text

pdf_path = "cssp_ibrahim.pdf"  # Replace with the path to your PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Step 3: Split the Extracted Text into Chunks
def split_text(text, chunk_size=500):
    text_chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    return text_chunks

pdf_chunks = split_text(pdf_text)

# Step 4: Create Embeddings for the Chunks
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
pdf_embeddings = embedding_model.encode(pdf_chunks)

# Step 5: Store the Embeddings in FAISS
import faiss
import numpy as np

dimension = pdf_embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(dimension)
faiss_index.add(np.array(pdf_embeddings))

# Step 6: Create a Retrieval Function and Generate Responses
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline  # Corrected import
from langchain.prompts import PromptTemplate

# Load your fine-tuned model for text generation
generate_pipeline = pipeline("text-generation", model="meta-llama/Llama-3.2-1B", tokenizer="meta-llama/Llama-3.2-1B")
llm = HuggingFacePipeline(pipeline=generate_pipeline)

# Define a prompt template
prompt_template = PromptTemplate(
    input_variables=["context", "query"],
    template="You are an expert assistant. Based on the following information:\n\n{context}\n\nAnswer the question: {query}"
)

def retrieve_similar_chunks(query, top_k=5):
    query_embedding = embedding_model.encode([query])
    distances, indices = faiss_index.search(np.array(query_embedding), top_k)
    return [pdf_chunks[idx] for idx in indices[0]]

def generate_rag_response(query):
    retrieved_chunks = retrieve_similar_chunks(query)
    context = " ".join(retrieved_chunks)  # Combine retrieved chunks
    input_prompt = prompt_template.format(context=context, query=query)  # Use the prompt template

    # Generate response with adjusted parameters
    response = llm(input_prompt, max_length=500)  # Increase the max_new_tokens limit
    return response[0]['generated_text']

# Example usage
query = "What is the main topic of the document?"
response = generate_rag_response(query)
print(response)
```
